```python
import keras
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
import pandas as pd
import json
import numpy as np
%matplotlib inline
```

```python
file_1 = 'data0.json'
file_2 = 'data1.json'
d1 = {}
d2 = {}
```

```python
with open(file_1,'r') as f:
    d1 = f.read()
```

```python
with open(file_2,'r') as f:
    d2 = f.read()
```

```python
d1 = json.loads(d1)
d2 = json.loads(d2)
```

```python
x = []
y = []
for i in range(len(d1['Ir_sensor'])):
    x.append([d1['Ir_sensor'][i],d1['Pin_sensor'][i]])
    if(i >= 43 and i <= 47):
        y.append(1)
    else:
        y.append(0)
for i in range(len(d2['Ir_sensor'])):
    x.append([d1['Ir_sensor'][i],d1['Pin_sensor'][i]])
    if(i >= 43 and i <= 47):
        y.append(1)
    else:
        y.append(0)
```

```python
np_x = np.array(x)
np_y = np.array(y)
```

```python
model = Sequential()
model.add(Dense(2,input_shape = (2,),kernel_initializer='normal', activation='relu'))
model.add(Dense(200,input_shape = (2,),kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=100,shuffle=True)
```
